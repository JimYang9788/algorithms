```python
class DLL:
    def __init__(self, val=0, key=None):
        self.val = val
        self.next = None 
        self.prev = None 
        self.key = key

class LRUCache:

    def __init__(self, capacity: int):
        
        self.cache = {}
        self.dlinkedlist = None 
        self.capacity = capacity
        
        self.size = 0
        self.head = DLL(0)
        self.tail = DLL(0)
    
        self.head.next = self.tail
        self.tail.prev = self.head 
        
        

    def get(self, key: int) -> int:
        node = self.cache.get(key)
        if not node:
            return -1 
        else:
            self._remove_node(node)
            self._create_at_front(node)
            return node.val
        
    def _create_at_front(self, node):
        node.prev = self.head
        node.next = self.head.next

        self.head.next.prev = node
        self.head.next = node
        # node.next, self.head.next.prev = self.head.next, node
        # node.prev, self.head.next =  self.head, node 
        # return True 
    
    def _remove_node (self, node):
        node.prev.next, node.next.prev = node.next, node.prev
        return node.key
        
    def _remove_last_node(self):
        # res = self.tail.prev
        # self._remove_node(res)
        # return res.key

        nodeKey = self.tail.prev.key
        # self.tail.prev.prev.next, self.tail.prev.next.prev = self.tail.prev.next, self.tail.prev.prev
        self.tail.prev.prev.next, self.tail.prev  = self.tail , self.tail.prev.prev
        return nodeKey

    def put(self, key: int, value: int) -> None:
        
        node = self.cache.get(key)
        if node:
            self._remove_node(node)
            self._create_at_front(node)
            self.cache[key].val = value
        
        else:
            node = DLL(value, key)
            self.cache[key] = node
            self._create_at_front(node)
            self.size += 1 
            if self.size  > self.capacity:
                removed_key = self._remove_last_node()

                del (self.cache[removed_key])
                self.size -= 1 
        
        return 
        


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
