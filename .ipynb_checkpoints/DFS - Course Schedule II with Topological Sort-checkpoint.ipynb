```python
class Solution:
    def findOrder(self, numCourses: int, prerequisites) :
        

        def dfs (i, order):
            
            if visited[i] == 2:
                return True
            
            if visited[i] == 1:
                return False
        
            visited[i] = 1 
            
            for neighbour in graph[i]:
                if not dfs (neighbour, order):
                    return False 
                
            order.append (i)
            visited[i] = 2 
            return True 
        # topological sort...
        
        graph = {i:[] for i in range(numCourses)}
        visited = {i: 0 for i in range (numCourses)}
        
        for edge in prerequisites:
            graph[edge[1]].append(edge[0])
            
        order = []
        
        for course in range(numCourses):
            if not dfs (course, order):
                return []
            
        order = order[::-1]
        return order 
        
```

```python
def topologicalSort(jobs, deps):
	
	
	
	def dfs (job, order):
		
		if visited[job] == 2:
			# Finished visited
			return True
		
		if visited[job] == 1:
			# being visited, found a loop
			return False 
		
		# start visiting the node
		visited[job] = 1 
		
		for neighbour in graph[job]:
			if not dfs (neighbour, order):
				return False 
		# finish visiting this node 
		visited[job] = 2 
		order.append (job)
		return True 
	
	graph = {job:[] for job in jobs}
	visited = {job: 0 for job in jobs}
	order = []

	for dep in deps:
		graph[dep[0]].append (dep[1])
		
	for job in jobs:
		if not dfs(job, order):
			return []
	
	return order[::-1]
```
