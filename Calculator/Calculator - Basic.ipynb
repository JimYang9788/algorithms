```python
class Solution:
    
    def calculate(self, s: str) -> int:
        # 每次见到一个新的符号，我们update的stack
        
        if not s:
            return 0 
        
        stack = []
        sign, num = '+', 0
        
        
        for i in range (len(s)):
            
            if s[i].isdigit ():
                num = num * 10 + int(s[i])
                
            if s[i] in '+-*/' or i == len(s) - 1:
                if sign == '+':
                    stack.append (num)
                elif sign == '-':
                    stack.append (-num)
                elif sign == '*':
                    stack.append (stack.pop() * num)
                elif sign == '/':
                    top = stack.pop()
                    if top >= 0:
                        stack.append (top // num)
                    elif top < 0:
                        stack.append (-((-top)//num))
                num = 0
                sign = s[i]
        return sum (stack)
```
