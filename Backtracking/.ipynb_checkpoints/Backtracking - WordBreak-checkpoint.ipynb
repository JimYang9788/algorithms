```python
# DFS Solution
# O(n^2)
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """
        hashmap = set()
        def backtrack (s, i, wordDict):
            
            if (i == len(s)): # Reach the end 
                return True 
            
            if i in hashmap:
                return False
            
            for word in wordDict:
                if s[i:].startswith(word):
                    length = len(word)
                    if backtrack(s, i + length,  wordDict):
                        return True
                    else:
                        hashmap.add(i+length)
                        
            return False 
            
        
        
        return backtrack (s, 0, wordDict)
```

```python
# DP Solution
class Solution(object):
    def wordBreak(self, s, wordDict):
        # DP Version 
        
        
        dp = [False for _ in range(len(s)+1)]
        dp[0] = True 
        
        for i in range(1,len(s)+1):
            for word in wordDict:
                length = len(word)
                if i-length >= 0 and dp[i - length] == True and s[i - length:i] == word:
                    dp[i] = True        
        return dp[len(s)]
```

```python
# BFS...? Stupid same shit as DFS 
class Solution:
    def wordBreak(self, s: str, wordDict) -> bool:
        word_set = set(wordDict)
        q = deque()
        visited = set()

        q.append(0)
        while q:
            start = q.popleft()
            if start in visited:
                continue
            for end in range(start + 1, len(s) + 1):
                if s[start:end] in word_set:
                    q.append(end)
                    if end == len(s):
                        return True
                visited.add(start)
        return False
```
