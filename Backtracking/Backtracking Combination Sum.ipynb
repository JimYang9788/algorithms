```python
class Solution:
    def combinationSum(self, candidates, target: int) :
        
        # An important detail on choosing the next number for the combination
        # is that we select the candidates in order, where the total candidates
        # are treated as a list. Once a candidate is added into the current combination,
        # we will not look back to all the previous candidates in the next explorations.
        def backtrack(candidates, n, temp,output,start):
            if n == 0:
                output.append (temp)
                return True
            
            # if n in visited:
            #     return False
                   
            res = False 
            
            for i in range(start, len(candidates)):
                if candidates[i] <= n and backtrack(candidates, n - candidates[i], temp+[candidates[i]],output,i):
                    res = True 
                # else:
                #     visited.add (n-candidates[i])         
            return res
        
        visited = set()
        seen = set()
        output = []
        
        backtrack (candidates, target, [], output,0)
        
        return output 
    
```

```python
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        
        def backtrack (output, candidates, n, idx, temp):
            
            if n == 0:
                output.append (temp)
                return 
            
            if n < 0:
                return   
            
            for i in range (idx, len(candidates)):
                if i == idx or candidates[i] != candidates[i-1]:
                    backtrack (output, candidates, n - candidates[i], i + 1, temp + [candidates[i]])        
        
        candidates.sort()
    
        output = []
        
        backtrack (output, candidates, target, 0, [])
        
        return output 
```
