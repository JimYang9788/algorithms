```python
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        
        def getCharacters(s):
            # Return a list of characters 
            res = set()
            for c in s:
                if c not in res:
                    res.add(c)
            # print (list(res))
            return list(res)
        
        def getAllPrefix(s):
            # Return a list of prefix
            res = []
            for i in range (0,len(s)+1):
                res.append (s[0:i])
            # print (res)
            return res 
        
        # Base Case 
        if s == "" and p == "":
            return True 

        elif s != "" and p == "":
            return False
        
        elif s == "" and p[0] != '*':
            return False
        

        elif p[0] == '?':
            for candidate in getCharacters(s):
                if self.isMatch(s, candidate + p[1:]):
                    return True
        elif p[0] == '*':
            for candidate in getAllPrefix(s):
                if self.isMatch(s, candidate + p[1:]):
                    return True
                
        elif s == "" and p != "":
            return False 
        
        else:        
            if s[0] == p[0] and self.isMatch(s[1:], p[1:]):

                return True 
            else:
                return False 
            
        
        
```

```python
class Solution:
# @return a boolean
    def isMatch(self, s, p):
        length = len(s)
        if len(p) - p.count('*') > length:
            return False
        dp = [True] + [False]*length
        for i in p:
            if i != '*':
                for n in reversed(range(length)):
                    dp[n+1] = dp[n] and (i == s[n] or i == '?')
            else:
                for n in range(1, length+1):
                    dp[n] = dp[n-1] or dp[n]
            dp[0] = dp[0] and i == '*'
        return dp[-1]
```

```python
class Solution:
    def isMatch(self, s, p):
        dp = [[False for _ in range(len(p)+1)] for i in range(len(s)+1)]
        dp[0][0] = True
        for j in range(1, len(p)+1):
            if p[j-1] != '*':
                break
            dp[0][j] = True
                
        for i in range(1, len(s)+1):
            for j in range(1, len(p)+1):
                if p[j-1] in {s[i-1], '?'}:
                    dp[i][j] = dp[i-1][j-1]
                elif p[j-1] == '*':
                    dp[i][j] = dp[i-1][j] or dp[i][j-1]
        return dp[-1][-1]
```
