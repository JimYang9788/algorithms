```python
class Solution:
    def calcEquation(self, equations, values, queries: List[List[str]]):
        
        
        # 1. build the graph using the edges 
        # 2. also build the  reverse edge using 
        # inverse of the weight
        
        
        # 3. Do a DFS. (we are trying to find a path, we don't care)
        # if it's the shortest path. In the DFS, we have to 
        # keep track of the current cost of reaching the node.
        
        graph = defaultdict (list)
        # {node: [(outnode, value)]} 
        
        for i, equation in enumerate (equations):
            outNode = equation[0]
            inNode = equation[1]
            
            graph[outNode].append ( (inNode, values[i]))
            graph[inNode].append((outNode, 1 / values[i]))
            
        res = []
        
        for query in queries:
            start = query[0]
            end = query[1]
            if not (start in graph and end in graph):
                res.append (-1)
            else:
                searchRes = self.dfs(start, end, set(),graph)
                res.append (searchRes)
        return res 
            
    def dfs (self, start, end, visited, graph):
        # return the cost of starting from start 
        # to reach the end 
        if start == end:
            return 1
        
        visited.add (start)
        
        for (neighbour, cost) in graph[start]:
            if neighbour not in visited:
                res = self.dfs (neighbour, end, visited, graph)
                if res != -1:                
                    return cost * res 
        return -1 
    
    
        
        
        
            
            
            
            
        
        
```
