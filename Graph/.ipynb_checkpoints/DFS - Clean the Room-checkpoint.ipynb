```python
# """
# This is the robot's control interface.
# You should not implement it, or speculate about its implementation
# """
#class Robot:
#    def move(self):
#        """
#        Returns true if the cell in front is open and robot moves into the cell.
#        Returns false if the cell in front is blocked and robot stays in the current cell.
#        :rtype bool
#        """
#
#    def turnLeft(self):
#        """
#        Robot will stay in the same cell after calling turnLeft/turnRight.
#        Each turn will be 90 degrees.
#        :rtype void
#        """
#
#    def turnRight(self):
#        """
#        Robot will stay in the same cell after calling turnLeft/turnRight.
#        Each turn will be 90 degrees.
#        :rtype void
#        """
#
#    def clean(self):
#        """
#        Clean the current cell.
#        :rtype void
#        """

class Solution:
    def cleanRoom(self, robot):
        """
        :type robot: Robot
        :rtype: None
        """
        def isAllCellVisited (visited):
            # Returns true if all the cell 
            # visited finished visited 
            for cell in visited.keys():
                if visited[cell] != 2:
                    return False
            return True
            
        def dfs (i,j, visited,robot, facing):
            robot.clean()
            
            if (i,j) in visited:
                robot.turnRight()
                robot.turnRight()
                robot.move()
                return 
            
            
            visited[(i,j)] = 1
            
            if isAllCellVisited (visited):
                return True 
            
            # Visit Up
            if facing == "up":
                if robot.move():
                    dfs (i,j+1,visited, robot, "up")
                    robot.turnRight()
                    robot.turnRight()

                # Visit Left
                robot.turnLeft()
                if robot.move():
                    dfs (i-1,j,visited, robot, "left")
                    robot.turnLeft()
                else:
                    robot.turnRight()

                # Visit Right
                robot.turnRight()
                if robot.move():        
                    dfs (i+1,j,visited, robot, "right")
                    robot.turnRight()
                else:
                    robot.turnLeft()

                # Visit Down 
                robot.turnRight()
                robot.turnRight()
                if robot.move():
                    dfs (i,j-1,visited, robot, "down")
                else:
                    robot.turnRight()
                    robot.turnRight()

            if facing == "right":
                if robot.move():
                    dfs (i+1,j,visited, robot, "right")
                    robot.turnRight()
                    robot.turnRight()
                    
                # Visit Left
                robot.turnLeft()
                if robot.move():
                    dfs (i,j+1,visited, robot, "up")
                    robot.turnLeft()
                else:
                    robot.turnRight()
                
                # Visit Right
                robot.turnRight()
                if robot.move():   
                    dfs (i,j-1,visited, robot, "down")
                    robot.turnRight()
                else:
                    robot.turnLeft()

                # Visit Down 
                robot.turnRight()
                robot.turnRight()
                if robot.move():
                    dfs (i-1,j,visited, robot, "left")
                else:
                    robot.turnRight()
                    robot.turnRight()

            if facing == "left":
                if robot.move():
                    dfs (i-1,j,visited, robot, "left")
                    robot.turnRight()
                    robot.turnRight()
                    
                # Visit Left
                robot.turnLeft()
                if robot.move():
                    dfs (i,j-1,visited, robot, "down")
                    robot.turnLeft()
                else:
                    robot.turnRight()

                # Visit Right
                robot.turnRight()
                if robot.move():        
                    dfs (i,j+1,visited, robot, "up")
                    robot.turnRight()
                else:
                    robot.turnLeft()

                # Visit Down 
                robot.turnRight()
                robot.turnRight()
                if robot.move():
                    dfs (i+1,j,visited, robot, "right")
                else:
                    robot.turnRight()
                    robot.turnRight()

            if facing == "down":
                if robot.move():
                    dfs (i,j-1,visited, robot, "down")
                    robot.turnRight()
                    robot.turnRight()

                # Visit Left
                robot.turnLeft()
                if robot.move():
                    dfs (i+1,j,visited, robot, "right")                
                    robot.turnLeft()
                else:
                    robot.turnRight()

                # Visit Right
                robot.turnRight()
                if robot.move():        
                    dfs (i-1,j,visited, robot, "left")
                    robot.turnRight()
                else:
                    robot.turnLeft()
                    
                # Visit Down 
                robot.turnRight()
                robot.turnRight()
                if robot.move():
                    dfs (i,j+1,visited, robot, "up")
                else:
                    robot.turnRight()
                    robot.turnRight()

            
            visited[(i,j)] = 2 
            robot.turnRight()
            robot.turnRight()
            robot.move()
            
        visited = {}
        # visited (i,j): 1 if being visited
        # visited (i,j): 2 if all neighbour finishes visited
        
        
        
        dfs (0,0,visited,robot, "up")
        
        
```
