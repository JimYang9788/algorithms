```python
class Solution:
    def pacificAtlantic(self, matrix: List[List[int]]) -> List[List[int]]:
        cache = {}
        
        def dfs (i, j, prev):
                
            if not (0 <= i < len(matrix) and 0 <= j < len(matrix[0])):
                return [0,0]
            
            if matrix[i][j] == '#':
                return [0,0]
        
            if prev < matrix[i][j]:
                return [0,0]
            
            if (i,j) in cache:
                return cache[(i,j)]
        
            temp = matrix[i][j]
            matrix[i][j] = '#'
            
            res = [0,0]
            if i == 0 or j == 0: res[0] = 1 
            if i == len(matrix) - 1 or j == len(matrix[0])-1: res[1] = 1 
            print (matrix)
            subRes = dfs(i-1,j, temp)
            res[0], res[1] = max(res[0],subRes[0]), max(res[1],subRes[1])

            subRes = dfs(i+1,j, temp)
 
            
            res[0], res[1] = max(res[0],subRes[0]), max(res[1],subRes[1])

            subRes = dfs(i,j-1, temp)
            res[0], res[1] = max(res[0],subRes[0]), max(res[1],subRes[1])

            subRes = dfs(i,j+1, temp)
            res[0], res[1] = max(res[0],subRes[0]), max(res[1],subRes[1])

            if i == 0 and j == 1:
                print (subRes)
            
            matrix[i][j] = temp
            
            cache[(i,j)] = res 
            return res
        
        output = []

        
        for i in range (0, len(matrix)):
            for j in range (0, len(matrix[0])):
                if dfs (i, j, float("inf")) == [1,1]:
                    output.append ([i,j])
        # print (cache)
        return output
    
```
