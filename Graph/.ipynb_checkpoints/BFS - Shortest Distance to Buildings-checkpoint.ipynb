```python
class Solution:
    def shortestDistance(self, grid):
        # A bruteforce approach 
        
        # run bfs on every single point, called start node,
        # in the graph
        # calculates the minimum distance between the start node
        # and the building. (this way we can skip 2, because we can't
        # expand).
        # after we find all the node, calculate the minimum distance. 
        
        
        # Keep track of all the buildings.
        # set of tuples 
        
        buildings = set()
        
        for i in range (len(grid)):
            for j in range (len(grid[0])):
                if grid[i][j] == 1:
                    buildings.add ((i,j))
                    
        
        
        def bfs (row,col):
            # returns the total distance to all the buildings 
            # starting at node i,j 
            totalDistance = 0
            
            queue = []
            queue.append ((row,col))
            visited = set()
            
            level = 0
            while(queue):
                if allVisited (visited, buildings):
                    break
                
                length = len (queue)
                for i in range (0, length):
                    i,j = queue.pop (0)
                    # 1. Check if Valid
                    if not (0 <= i < len(grid) and 0 <= j < len(grid[0])):
                        continue 
                        
                    # 1.5 Check if visited:
                    if (i,j) in visited:
                        continue
                    
                    # 2. Check if it's a buiding
                    elif grid[i][j] == 1:
                        totalDistance += level
                        visited.add ((i,j))
                        continue 
                    
                    # 3. Check if it's a obstacle 
                    elif grid[i][j] == 2:
                        continue 
                    
                    for neighbour in getNeighbour (i,j):
                        queue.append (neighbour)
                    
                    visited.add ((i,j))
                    
                level += 1 
                    
            if not allVisited (visited, buildings):
                return float("inf")
            else:
                return totalDistance
        
        def getNeighbour (i,j):
            return [(i+1,j), (i-1,j) ,(i,j+1), (i,j-1)]
        
        
        def allVisited (visited, buildings):
            for building in buildings:
                if building not in visited:
                    return False 
            return True 
        

        if not grid and not grid[0]:
            return 0
        
        
        minimumDistance = float ("inf")
        for i in range (len(grid)):
            for j in range (len(grid[0])):
                if grid[i][j] == 0:
                    minimumDistance = min (minimumDistance, bfs (i,j))
        
        return minimumDistance if minimumDistance != float("inf") else -1 
                    
```

```python
class Solution:
    def shortestDistance(self, grid):
        # A bruteforce approach 
        
        # run bfs on every single point, called start node,
        # in the graph
        # calculates the minimum distance between the start node
        # and the building. (this way we can skip 2, because we can't
        # expand).
        # after we find all the node, calculate the minimum distance. 
        
        
        # Keep track of all the buildings.
        # set of tuples 
        
        buildings = set()
        
        for i in range (len(grid)):
            for j in range (len(grid[0])):
                if grid[i][j] == 1:
                    buildings.add ((i,j))
                    
        hit = [[0 for col in range (len(grid[0])) ] for row in grid]
        distSum = [[0 for col in range (len(grid[0])) ] for row in grid]

        
        def bfs (row,col):

            
            queue = []
            
            for neighbour in getNeighbour (row,col):
                queue.append (neighbour)
                
            visited = set ()
            visited.add ((row,col))
            count = 1 
            level = 0 
            
            while(queue):
                length = len (queue)
                for i in range (0, length):
                    i,j = queue.pop (0)
                    # 1. Check if Valid
                    if not (0 <= i < len(grid) and 0 <= j < len(grid[0])):
                        continue 
                        
                    # 1.5 Check if visited:
                    if (i,j) in visited:
                        continue
                    
                    # 2. Check if it's a buiding
                    elif grid[i][j] == 1:
                        visited.add ((i,j))
                        count += 1 
                        continue 
                        
                    elif grid[i][j] == 0:
                        hit [i][j] += 1 
                        distSum[i][j] += level + 1 
                        
                    # 3. Check if it's a obstacle 
                    elif grid[i][j] == 2:
                        continue 
                    
                    for neighbour in getNeighbour (i,j):
                        queue.append (neighbour)
                    
                    visited.add ((i,j))
                level += 1 
            return count == len(buildings)
        
        def getNeighbour (i,j):
            return [(i+1,j), (i-1,j) ,(i,j+1), (i,j-1)]
        
        
        def allVisited (visited, buildings):
            for building in buildings:
                if building not in visited:
                    return False 
            return True 
        

        if not grid and not grid[0]:
            return 0
        
        
        minimumDistance = float ("inf")
        for i in range (len(grid)):
            for j in range (len(grid[0])):
                if grid[i][j] == 1:
                    if not bfs (i,j): return -1 
        
        for i in range (len(hit)):
            for j in range (len(hit[0])):
                if grid[i][j] == 0:
                    if hit[i][j] == len(buildings):
                        minimumDistance = min (distSum[i][j], minimumDistance)
        
        print (distSum)
        return minimumDistance if minimumDistance != float("inf") else -1 
```
