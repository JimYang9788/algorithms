```python
class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        # BFS 
        # Optimization, using double BFS 
        
        def isAdjacent(curword, word):
            diff = 0 
            i = 0
            while (i < len(curword) and i < len(word)):
                if diff > 1:
                    return False
                if curword[i] != word[i]:
                    diff += 1 
                i += 1 
            
            if i < len(curword):
                diff += len(curword)
            if i < len(word):
                diff += len (word)
            
            if diff >1:
                return False 
            else:
                return True 
            
        def getNeighbour (curword, wordList):
            # Return list of the neighbours of curword in the wordList 
            res = []
            for word in wordList:
                if isAdjacent (curword, word):
                    res.append (word)
            return res 
        
        queue = set()
        queue2 = set()
        visited = set()
        
        graph = {word: [] for word in wordList + [beginWord] + [endWord]}
        
        for word in graph.keys():
            for potentialNeighbour in wordList:
                if isAdjacent (word, potentialNeighbour):
                    graph[word].append(potentialNeighbour)
                    

        
        queue.add (beginWord)
        queue2.add(endWord)
        step = 1
        
        while (queue):
            temp = set ()
            
            for curword in queue:
                
                if curword in queue2:
                    return step 
                
                visited.add (curword)
                
                for neighbour in graph[curword]:
                    if neighbour not in visited:
                        temp.add (neighbour)
                        
            queue = queue2 
            queue2 = temp 
            
            step += 1 
            
        return 0 
                    
            
        
        
```
