```python
class LFUCache:

    def __init__(self, capacity: int):
        self.capacity = capacity 
        self.key_value_map = defaultdict (int)
        self.key_frequency_map = defaultdict (int)
        self.freq_to_key = defaultdict (OrderedDict)
        self.min_freq = 0

    def get(self, key: int) -> int:
        if key not in self.key_value_map:
            return -1 
        self.increaseFrequency(key)
        # print ('get ', key,  self.key_value_map, self.key_frequency_map, self.freq_to_key)
        return self.key_value_map[key]

    
    def increaseFrequency(self, key):
        self.key_frequency_map[key] += 1 
        self.freq_to_key [self.key_frequency_map[key] - 1].pop(key)
        if self.min_freq == self.key_frequency_map[key] - 1 and len(self.freq_to_key[self.min_freq].keys()) == 0:
            self.min_freq = self.min_freq + 1 
        self.freq_to_key [self.key_frequency_map[key]][key] = 1
    
    
    def removeMinFreqKey(self):
        del_key, val = self.freq_to_key[self.min_freq].popitem(False)
        del self.key_value_map[del_key]
        del self.key_frequency_map[del_key]
        if len(self.freq_to_key[self.min_freq]) == 0:
            del self.freq_to_key[self.min_freq]
            
    
    def put(self, key: int, value: int) -> None:
        if self.capacity <= 0: return 
        
        if key in self.key_value_map:
            self.increaseFrequency(key)
            self.key_value_map[key] = value 
            return 
        
        if len(self.key_value_map.keys()) >= self.capacity:
            self.removeMinFreqKey()
        
        self.key_value_map[key] = value 
        self.min_freq = 1 # a new key is inserted
        self.key_frequency_map[key] = 1 
        if self.min_freq not in self.freq_to_key:
            self.freq_to_key[self.min_freq] = OrderedDict ()
        self.freq_to_key[self.min_freq][key] = 1
        
        # print ('put ', key, self.key_value_map, self.key_frequency_map, self.freq_to_key)
        return 

# Your LFUCache object will be instantiated and called as such:
# obj = LFUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
