```python
# https://www.youtube.com/watch?v=YcJTyrG3bZs&ab_channel=BackToBackSWE
# Far better! 

class Solution:
    def numDecodings(self, s: str) -> int:
        
        # check content validity
        def valid(s):
            if s[0] == '0' or len(s) == 0:
                return False
            return int(s)>=1 and int(s)<=26
        
        # 123
        def inner(s,ptr,dp):
            
            # found exact match(ptr went to very end)
            if ptr==len(s):
                return 1
            
            # previously cached result
            if dp[ptr] != -1:
                return dp[ptr]
                
            
            # calculate total from 
            # 1 or 2 characters selected
            total = 0
            for i in range(1,3):
                if ptr+i <= len(s):
                    snip = s[ptr:ptr+i]
                    
                    if valid(snip):
                        total += inner(s,ptr+i,dp)
                
            
            dp[ptr] = total
            return dp[ptr]
            
            
        
        n = len(s)
        dp = [-1 for _ in range(n)]
        
        return inner(s,0,dp)
```
