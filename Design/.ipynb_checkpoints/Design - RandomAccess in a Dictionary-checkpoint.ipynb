```python
class RandomizedSet:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        
        
        self.array = []
        self.hashmap = {}

    def insert(self, val: int) -> bool:
        """
        Inserts a value to the set. Returns true if the set did not already contain the specified element.
        """
        if val in self.hashmap:
            return False 
        else:
            self.array.append (val)
            self.hashmap[val] = len(self.array) - 1  
            return True 

    def remove(self, val: int) -> bool:
        """
        Removes a value from the set. Returns true if the set contained the specified element.
        """
        if val not in self.hashmap:
            return False 
        else:
            # We want to delete this guy 
            
            targetidx = self.hashmap [val]

            
            # Swap the last element and the target in the hashmap
            self.hashmap[val], self.hashmap[self.array[-1]] = self.hashmap[self.array[-1]], self.hashmap[val] 
            # Swap the last element and the target in the array
            self.array[targetidx], self.array[len(self.array)-1] = self.array[len(self.array)-1],self.array[targetidx]

            
            del self.array[len(self.array)-1]
            del self.hashmap[val]
            
            return True 
        

    def getRandom(self) -> int:
        """
        Get a random element from the set.
        """
        r = random.randrange(0, len(self.array), 1)
        return self.array[r]
        
        
        


# Your RandomizedSet object will be instantiated and called as such:
# obj = RandomizedSet()
# param_1 = obj.insert(val)
# param_2 = obj.remove(val)
# param_3 = obj.getRandom()
```
