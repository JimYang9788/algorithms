```python
from collections import defaultdict 
class Transaction:
    def __init__(self,s):
        
        data = s.split (',')
        assert (len(data) == 4)

        self.name = data[0] 
        self.time = int (data[1])
        self.amount = int (data[2])
        self.city = data[3]
        self.invalid = False 
    
    def __lt__(self, other):
        return self.time < other.time 

class Solution:
    def invalidTransactions(self, transactions):
        '''
        sort transaction by start time  

        for transaction i transactions:
            push transaction to the tmap, detect if there's a invalid transaction 
            pop every single transaction that's within 60 minutes of interval and push to invalid 
        
            ## Optimization after 60 minutes we can clean the queue by some mechanism 
        '''
        transactions = list(map (lambda t: Transaction (t), transactions)) 
        transactions.sort()
        
        invalid = []
        tmap  = defaultdict(list)
        
        for t in transactions:
            time = t.time 
            name = t.name 
            city = t.city 
            amount = t.amount 
            
            if amount > 1000:
                t.invalid = True 

            i = len(tmap[name]) - 1 
            while (i >= 0 and abs (tmap[name][i].time - time) <= 60 and tmap[name][i].city != city):
                tmap[name][i].invalid = True 
                t.invalid = True 
                i -= 1 
                
            tmap[name].append (t)
            
        for name, lot in tmap.items():
            lot.sort()
#             print (name, list((t.name, t.time, t.amount, t.invalid, t.city) for t in lot))
            for t in lot:
                
                if t.invalid:
                    invalid.append (t.name + ',' + str (t.time) + "," + str (t.amount) + "," + t.city)
        
        return invalid 
    
    def parse (self, transactions):

        transactions = list(map (lambda t: Transaction (t), transactions)) 
        transactions.sort()

        
        invalid = []
        tmap  = defaultdict(list)
        for t in transactions:
            tmap[t.name].append (t)
        
#         for name, lot in tmap.items():
#             print (name, [(t.name, t.time, t.amount, t.city) for t in lot])
                
        return tmap 
```
