#### Q2 Middle of the Linked list

Given a non-empty, singly linked list with head node head, return a middle node of linked list.

If there are two middle nodes, return the second middle node.

````python
class Solution(object):
    def middleNode(self, head):
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow
````

#### Q3 Backstring space compare 
Given two strings S and T, return if they are equal when both are typed into empty text editors. # means a backspace character.

```
Input: S = "ab#c", T = "ad#c"
Output: true
Explanation: Both S and T become "ac".
```
\# is a backspace character


````python
class Solution:
    # O(N) Memory is trivial. Let's try O(1) Memory
    
    def remove_backspace (self,S):
        S = S[::-1]
        skip_counter = 0
        res = ''
        for i in range (len(S)):
            if skip_counter != 0:
                if S[i] == '#':
                    skip_counter += 1 
                else:
                    skip_counter -= 1 
            else:
                if S[i] == '#':
                    skip_counter += 1 
                else:
                    res = S[i] + res 
        return res
        
    
    def backspaceCompare(self,S,T):
        resS = self.remove_backspace (S)
        resT = self.remove_backspace (T)
        return resS == resT

solution = Solution()
solution.backspaceCompare("ab##","c#d#")
                
# Better Solution

class Solution(object):
    def backspaceCompare(self, S, T):
        def F(S):
            skip = 0
            for x in reversed(S):
                if x == '#':
                    skip += 1
                elif skip:
                    skip -= 1
                else:
                    yield x # Returns a generator 

        return all(x == y for x, y in itertools.izip_longest(F(S), F(T)))
    # itertools returns (p[0], q[0]), (p[1], q[1]), …
````

#### Q4 Intersection of two arrays 
Given two arrays, write a function to compute their intersection.

Example 1:

Input: nums1 = [1,2,2,1], nums2 = [2]
Output: [2]

````python
class Solution:
    def bin_search (self, A, n):
        L, U = 0, len(A)-1
        while (L <= U):
            M = (L+U)//2
            if A[M] < n:
                L = M + 1 
            elif A[M] > n:
                U = M - 1
            else:
                return True
        return False
        
    def intersect(self, nums1, nums2):
        nums1.sort() # O(nlogn)
        nums2.sort() # O(nlogn)
        res = {}
        for i in range (len(nums1)):
            if self.bin_search(nums2, nums1[i]):
                if nums1[i] in res:
                    res[nums1[i]] += 1 
                else:
                    res[nums1[i]] = 1 
        res2 = {} 
        for i in range (len(nums2)):
            if self.bin_search(nums1, nums2[i]):
                if nums2[i] in res2:
                    res2[nums2[i]] += 1 
                else:
                    res2[nums2[i]] = 1    
        final_res = []
        
        for key,val in res.items():
            final_res += [key] * min(res[key],res2[key])
        
        return final_res 
Solution().intersect([1,3,3], [2,2,3])
# assert (Solution().intersect([1,2,2,3], [2,2,3]) == [2,2,3])
````

Solution Notes:

##### Approach 1: Hash Map
For the previous problem, we used a hash set to achieve a linear time complexity. Here, we need to use a hash map to track the count for each number.

We collect numbers and their counts from one of the arrays into a hash map. Then, we iterate along the second array, and check if the number exists in the hash map and its count is positive. If so - add the number to the result and decrease its count in the hash map.
用小的array建立hashmap，然后检查在不在hashmap里面，在的话counter-=1然后放进result

##### Approach 2: Sorting + Lock Bottle Neck Proceed
Sort nums1 and nums2.

Initialize i, j and k with zero.

Move indices i along nums1, and j through nums2:

Increment i if nums1[i] is smaller.

Increment j if nums2[j] is smaller.

If numbers are the same, copy the number into nums1[k], and increment i, j and k.

Return first k elements of nums1.

```
public int[] twoSum(int[] nums, int target) {
    Map<Integer, Integer> map = new HashMap<>();
    for (int i = 0; i < nums.length; i++) {
        int complement = target - nums[i];
        if (map.containsKey(complement)) {

    }
    throw new IllegalArgumentException("No two sum solution");
```

````python
from collections import Counter
from collections import defaultdict  

class Solution:
    def twoSum(self, nums):
        hash_map = defaultdict (int)

        for idx, val in enumerate (nums):
            complement = target - val 
            if complement in hash_map:
                return [hash_map[complement], idx]
            else:
                hash_map [val] = idx
# 不需要two pass，同时巧妙的避开了duplicate，因为duplicate只有2*duplicate = target的时候我们才care，
# 那么可以在加入hash map之前就做duplicate的comparison        
````

#### Q6 MinStack Implementation 


````python
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.minstack = []
        self.stack = []

    def push(self, x):
        if self.minstack == [] and self.stack == []:
            self.minstack = [x] 
            self.stack = [x]
        else:
            self.stack.append (x)
            if x < self.minstack[-1]:
                self.minstack.append (x)
            else:
                self.minstack.append (self.minstack[-1])
            
    def pop(self):
        self.stack.pop ()
        self.minstack.pop ()

    def top(self):
        return self.stack[-1]

    def getMin(self):
        return self.minstack[-1]


# Your MinStack object will be instantiated and called as such:
obj = MinStack()
obj.push(3)
obj.push(4)
obj.push(4)
obj.push(4)
obj.push(4)
obj.pop()
obj.stack
obj.getMin() # O(1) retrieve smallest element
````

#### Diameter of a binary tree
Given a binary tree, you need to compute the length of the diameter of the tree. The diameter of a binary tree is the length of the longest path between any two nodes in a tree. This path may or may not pass through the root.

```
          1
         / \
        2   3
       / \     
      4   5    
```
The longest path here is [4,2,1,3]

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import namedtuple 
NodeInfo = namedtuple('NodeInfo', ['height', 'max_diameter'])

class Solution:
    def getNodeInfo (self, root):
        if root == None: return NodeInfo(0,0)
        
        left_node_info = self.getNodeInfo(root.left)
        right_node_info = self.getNodeInfo(root.right)
        
        current_height = max(left_node_info.height,right_node_info.height)+1
        current_max_diameter = max(left_node_info.height + right_node_info.height, 
                                   left_node_info.max_diameter, right_node_info.max_diameter)
        
        return NodeInfo (current_height,current_max_diameter) 
        
        
    def diameterOfBinaryTree(self, root):
        
        return self.getNodeInfo (root).max_diameter

# 标准答案用了一个global variable self.ans来记录当前最大
# space上比我的好(我每次call stack要多return一个variable)，
# 虽然space同时都要O(N)的call stack

class Solution(object):
    def diameterOfBinaryTree(self, root):
        self.ans = 1
        def depth(node):
            if not node: return 0
            L = depth(node.left)
            R = depth(node.right)
            self.ans = max(self.ans, L+R+1)
            return max(L, R) + 1

        depth(root)
        return self.ans - 1
    
# NodeInfo = namedtuple('NodeInfo', ['height', 'max_diameter'])
# NodeInfo(1,1).height => 1 
````

#### Valid Binary Search Tree

Given a binary tree, determine if it is a valid binary search tree (BST).

Assume a BST is defined as follows:

The left subtree of a node contains only nodes with keys less than the node's key.
The right subtree of a node contains only nodes with keys greater than the node's key.
Both the left and right subtrees must also be binary search trees.


````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

NodeInfo = namedtuple ('NodeInfo',['is_valid','min','max'])

class Solution:
    def getNodeInfo (self, node):
        if node == None:
            return NodeInfo (True,None,None)
        elif node.left == None and node.right == None:
            return NodeInfo (True,node.val,node.val)
        
        right_node_res = self.getNodeInfo (node.right)
        left_node_res = self.getNodeInfo (node.left)
        
        cur_valid = right_node_res.is_valid and left_node_res.is_valid  
        
        cur_min = None 
        cur_max = None 
        
        if not node.right:
            cur_min = min (left_node_res.min, node.val)
            cur_max = max (left_node_res.max, node.val)
            cur_valid = cur_valid and node.val > left_node_res.max 

        elif not node.left:
            cur_min = min (right_node_res.min, node.val)
            cur_max = max (right_node_res.max, node.val)
            cur_valid = cur_valid and node.val < right_node_res.min 
    
        else:
            cur_min = min (left_node_res.min, right_node_res.min, node.val)
            cur_max = max (left_node_res.max, right_node_res.max, node.val)
            cur_valid = cur_valid and node.val < right_node_res.min and node.val > left_node_res.max 

        return NodeInfo (cur_valid, cur_min, cur_max)
            
        
    def isValidBST(self, root):
        return self.getNodeInfo(root).is_valid 
        

### 标准答案直接用了负无限大...
class Solution:
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        def helper(node, lower = float('-inf'), upper = float('inf')):
            if not node:
                return True
            
            val = node.val
            if val <= lower or val >= upper:
                return False

            if not helper(node.right, val, upper):
                return False
            if not helper(node.left, lower, val):
                return False
            return True

        return helper(root)
    
## Iteration! 
class Solution:
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root:
            return True
            
        stack = [(root, float('-inf'), float('inf')), ] 
        while stack:
            root, lower, upper = stack.pop()
            if not root:
                continue
            val = root.val
            if val <= lower or val >= upper:
                return False
            stack.append((root.right, val, upper))
            stack.append((root.left, lower, val))
        return True  
````
