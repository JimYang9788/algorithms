```python
import heapq

class ContinuousMedianHandler:
    def __init__(self):
        self.median = None
        self.minHeap = [] 
        self.maxHeap = []

    def insert(self, number):
        if not self.minHeap:
            heapq.heappush(self.minHeap, number)
            return
        if not self.maxHeap:
            heapq.heappush(self.maxHeap, -number)
            return

        greater = heapq.heappop(self.minHeap)
        smaller = -heapq.heappop(self.maxHeap)

        if len(self.maxHeap) == len(self.minHeap):
            insert = sorted ([greater, smaller, number])
            heapq.heappush(self.maxHeap, -insert[0])
            heapq.heappush(self.maxHeap, -insert[1])
            heapq.heappush(self.minHeap, insert[2])

        elif len(self.maxHeap) > len(self.minHeap):
            # determine which number to go to minheap 

            if number > smaller:
                heapq.heappush(self.minHeap, number)
                heapq.heappush(self.maxHeap, -smaller)
            else:
                heapq.heappush(self.minHeap, smaller)
                heapq.heappush(self.maxHeap, -number)				
            heapq.heappush (self.minHeap, greater)
        else:
            # determine which number to go to maxheap 
            if number > greater:
                heapq.heappush (self.maxHeap, -greater)
                heapq.heappush (self.minHeap, number)
            else:
                heapq.heappush (self.maxHeap, -number)
                heapq.heappush (self.minHeap, greater)			
            heapq.heappush (self.maxHeap, -smaller)


    def getMedian(self):
        if not self.minHeap and not self.maxHeap:
            return None
        elif not self.minHeap:
            return self.maxHeap[0]
        elif not self.maxHeap:
            return self.minHeap[0]

        if len(self.minHeap) == len(self.maxHeap):
            return (self.minHeap[0] + -self.maxHeap[0])/2 

        elif len(self.minHeap) > len(self.maxHeap):
            return self.minHeap[0]
        else:
            return -self.maxHeap[0]
```
