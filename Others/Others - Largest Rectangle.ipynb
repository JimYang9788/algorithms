```python
class Solution:
    def largestRectangleArea(self, heights) -> int:
        # Brute Force
        # O(n^3)
        n = len(heights)
        maxArea = 0
        for i in range(0,n):
            for j in range (i+1, n+1):
                minNum = min(heights[i:j])
                area = minNum * (j-i)
                maxArea = max(area, maxArea)
        
        return maxArea 
        
```

```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        
        cache = [0 for _ in heights]
        
        def leftScan (i, minHeight):
            total = 0 
            tbCache = []
            while (i >= 0):
                if heights[i] == minHeight:
                    tbCache.append(i)
                    total += minHeight

                elif heights[i] < minHeight:
                    break
                else:
                    total += minHeight
                i -= 1 
            return total, tbCache
    
        def rightScan (i, minHeight):
            total = 0 
            tbCache = []
            while (i <= len(heights)-1):
                if heights[i] == minHeight:
                    tbCache.append(i)
                    total += minHeight
                elif heights[i] < minHeight:
                    break
                else:
                    total += minHeight
                i += 1 
            return total  , tbCache       
        
        # O(n^2)
        maxArea = 0 
        for i in range (0, len(heights)):
            if cache[i]:
                maxArea = max(cache[i], maxArea)
            else:
                curArea = heights[i] 
                leftArea, leftCache = leftScan (i-1,heights[i])
                rightArea, rightCache = rightScan (i+1, heights[i])
                newCache = leftCache + rightCache
                newArea = leftArea + rightArea + curArea
                for index in newCache:
                    cache[index] = newArea
                maxArea = max (newArea, maxArea)
        
        return maxArea 
```
