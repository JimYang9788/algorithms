```python
# Linked List EvenOdd Head 
class Solution:
    def oddEvenList(self, head):
        
        odd = oddhead = ListNode ()
        even = evenhead = ListNode ()
        
        is_odd = True 
        
        while (head):    
            if is_odd:
                odd.next = head
                odd = odd.next 
            else:
                even.next = head 
                even = even.next 
                
            is_odd = not is_odd
            head = head.next 
        
        
        odd.next = evenhead.next
        even.next = None 
        
        return oddhead.next 
```

```python

class Solution:

    
    def zigzagLevelOrder(self, root: TreeNode):
        if not root: return []
        
        res = []
        Q = []
        Q.append (root)
        left_to_right = 1 
        while Q:
            next_level = []
            for node in Q:
                if node.left:
                    next_level.append (node.left)
                if node.right:
                    next_level.append (node.right)
            
            if left_to_right:
                res.append ([node.val for node in Q])
            else:
                Q.reverse()
                res.append ([node.val for node in Q])
            left_to_right = 1 - left_to_right
            
            Q = next_level
        return res
            
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

        
    def buildTree(self, preorder, inorder):
        
        if not inorder: return None
        
        mid = preorder[0]
        mid_pos = inorder.index (mid)
        
        treenode = TreeNode (mid,None,None)
        treenode.left = self.buildTree(preorder[1:1+mid_pos],inorder[0:mid_pos])
        treenode.right = self.buildTree(preorder[1+mid_pos:],inorder[mid_pos+1:])
        
        return treenode
    
```

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        mapping = {
            2: ['a', 'b', 'c'],
            3: ['d','e', 'f'],
            4: ['g','h', 'i'],
            5: ['j','k', 'l'],
            6: ['m','n', 'o'],
            7: ['p','q', 'r','s'],
            8: ['t','u', 'v'],
            9: ['w','x', 'y','z']
        }
        if digits == "": return []
        res  = [""]
        for c in digits:
            input_list = mapping [int(c)]
            new_results = []
            for word in res:
                new_word_list = [word + suffix for suffix in input_list]
                new_results += new_word_list
            res = new_results 
        return res 
```

```python
class Solution(object):
    def generateParenthesis(self, N):
        ans = []
        def backtrack(S = '', left = 0, right = 0):
            if len(S) == 2 * N:
                ans.append(S)
                return
            if left < N:
                backtrack(S+'(', left+1, right)
            if right < left:
                backtrack(S+')', left, right+1)

        backtrack()
        return ans
```

```python
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        if not nums: return []
        if len(nums) == 1: return [nums]
        
        prev_res = self.permute (nums[1:])
        res = []
        n = nums[0]
        for perm in prev_res:
            for i in range (0, len(perm)+1):
                new_perm = perm.copy()
                new_perm.insert(i, n)
                res.append (new_perm)
        return res 
```

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        if not nums: return []
        if len(nums)==1: return [nums,[]] 
        res,n = [],nums[0]
        prev_res = self.subsets (nums[1:])
        for subset in prev_res:
            new_subset = subset.copy()
            new_subset.append (n)
            res.append(subset)
            res.append(new_subset)
        return res
            
```

```python
class Solution(object):
    def exist(self, board, word):
        """
        :type board: List[List[str]]
        :type word: str
        :rtype: bool
        """
        self.ROWS = len(board)
        self.COLS = len(board[0])
        self.board = board

        for row in range(self.ROWS):
            for col in range(self.COLS):
                if self.backtrack(row, col, word):
                    return True

        # no match found after all exploration
        return False, row, col, suffix
        # bottom case: 


    def backtrack(self):
        if len(suffix) == 0:
            return True 

        # Check the current status, before jumping into backtracking
        if row < 0 or row == self.ROWS or col < 0 or col == self.COLS \
                or self.board[row][col] != suffix[0]:
            return False

        ret = False
        # mark the choice before exploring further.
        self.board[row][col] = '#'
        # explore the 4 neighbor directions
        for rowOffset, colOffset in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            ret = self.backtrack(row + rowOffset, col + colOffset, suffix[1:])
            # break instead of return directly to do some cleanup afterwards
            if ret: break

        # revert the change, a clean slate and no side-effect
        self.board[row][col] = suffix[0]

        # Tried all directions, and did not find any match
        return ret
```

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Dutch National Flag problem solution.
        """
        # for all idx < p0 : nums[idx < p0] = 0
        # curr is an index of element under consideration
        p0 = curr = 0
        # for all idx > p2 : nums[idx > p2] = 2
        p2 = len(nums) - 1

        while curr <= p2:
            if nums[curr] == 0:
                nums[p0], nums[curr] = nums[curr], nums[p0]
                p0 += 1
                curr += 1
            elif nums[curr] == 2:
                nums[curr], nums[p2] = nums[p2], nums[curr]
                p2 -= 1
            else:
                curr += 1
```

```python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count_list = {}
        for i in nums:
            if i in count_list:
                count_list[i] +=  1 
            else:
                count_list[i] = 1
        h = []
        for num,freq in count_list.items():
            heappush(h, (-freq, num))
        res = []
        while k != 0:
            (freq,num) = heappop (h)
            res.append (num)
            k -= 1
        return res 
```

```python
class Solution:
    def shortestDistance(self, words, word1, word2):
        size = len(words)
        index1, index2 = size, size
        ans = size

        for i in range(size):
            if words[i] == word1:
                index1 = i
                ans = min(ans, abs(index1-index2))
            elif words[i] == word2:
                index2 = i
                ans = min(ans, abs(index1-index2))
        return ans
        
```

```python
def canJump(self, nums):
    max_reach, n = 0, len(nums)
    for i, x in enumerate(nums):
        if max_reach < i: return False
        if max_reach >= n - 1: return True
        max_reach = max(max_reach, i + x)
```
