```python
class TicTacToe:

    def __init__(self, n: int):
        """
        Initialize your data structure here.
        """
        assert(n>0)
        self.board = []
        for i in range (n):
            self.board.append (list (0 for i in range (n)))
        self.winner = 0
        self.size = n
        
    def print_board (self):
        for row in self.board:
            for val in row:
                print (val)
            print ("\n")

        
    def move(self, row: int, col: int, player: int) -> int:
        """
        Player {player} makes a move at ({row}, {col}).
        @param row The row of the board.
        @param col The column of the board.
        @param player The player, can be either 1 or 2.
        @return The current winning condition, can be either:
                0: No one wins.
                1: Player 1 wins.
                2: Player 2 wins.
        """
        def advance_check (row,col,player):
            def is_same (val, nums):
                for num in nums:
                    if num != val:
                        return False
                return True
            if is_same(player,list(self.board[row][i] for i in range (self.size))):
                self.winner = player  
            elif is_same(player,list(self.board[i][col] for i in range (self.size))):
                self.winner = player  
            elif row == col or row == self.size - 1 - col:
                if is_same(player,list(self.board[i][self.size-1-i] for i in range(self.size))):
                    self.winner = player  
                if is_same(player,list(self.board[i][i] for i in range(0,self.size))):
                    self.winner = player  


                

        if self.winner != 0: return self.winner
        self.board[row][col] = player 

        if advance_check (row, col,player): self.winner = player        

        return self.winner 
        
        
        
        


# Your TicTacToe object will be instantiated and called as such:
# obj = TicTacToe(n)
# param_1 = obj.move(row,col,player)
```
