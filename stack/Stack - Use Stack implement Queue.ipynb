```python
class Stack:
    def __init__(self):
        self.stack = []
        
    def push (self, x):
        self.stack.append(x)
    
    def pop(self):
        if self.stack:
            self.stack.pop()
            return True
        else:
            return False
    
    def top(self):
        return self.stack[-1]
    
    def empty(self):
        return len (self.stack) == 0

class MyQueue:

    def __init__(self):
        self.s1 = Stack()
        self.s2 = Stack()

    def push(self, x: int) -> None:
        self.s1.push(x)

    def pop(self) -> int:
        while not self.s1.empty():
            self.s2.push(self.s1.top())
            self.s1.pop()
        element = self.s2.top()
        self.s2.pop()
        
        while not self.s2.empty():
            self.s1.push(self.s2.top())
            self.s2.pop()   
        return element

    def peek(self) -> int:
        while not self.s1.empty():
            self.s2.push(self.s1.top())
            self.s1.pop()
            
        element = self.s2.top()
        
        while not self.s2.empty():
            self.s1.push(self.s2.top())
            self.s2.pop()   
        return element
        
    def empty(self) -> bool:
        return self.s1.empty()


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
